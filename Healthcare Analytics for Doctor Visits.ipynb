```python
sns.histplot(df.gender, bins = 2);
```

```python
label = ["yes","no"]
y = df[df["freepoor"] == "yes"]
n = df[df["freepoor"] == "no"]
x = [y.shape[0], n.shape[0]]

plt.figure(figsize = (5,5))
plt.pie(x,labels = label)
plt.title("% of people getting govt heath insurance due to low income")

plt.show()
```

```python
y = df[df["private"] == "yes"]
n = df[df["private"] == "no"]
x = [y.shape[0], n.shape[0]]

plt.figure(figsize = (5,5))
plt.pie(x,labels = label)
plt.title("% of people having private insurance ")

plt.show()
```

```python
y = df[df["freerepat"] == "yes"]
n = df[df["freerepat"] == "no"]
x = [y.shape[0], n.shape[0]]

plt.figure(figsize = (5,5))
plt.pie(x,labels = label)
plt.title("% of people getting govt heath insurance due to old age, disability or veteran status")

plt.show()
```

```python
db = df.groupby("gender") ["reduced"].sum().to_frame().reset_index()
```

```python
plt.barh(db["gender"], db["reduced"], color = ["cornflowerblue","lightseagreen"])
plt.title("Bar Chart")
plt.xlabel("gender")
plt.ylabel("reduced activity")

plt.show()
```
